```r
library(forecast)
library(ggplot2)

set.seed(12)

y <- ts(numeric(300))
e <- rnorm(300)

# 0.9, 0.6, -0.8, -0.5
for (t in 2:300){
    y[t] <- 0.9*y[t-1]+e[t]  # ø
}

ggtsdisplay(y)
Arima(y, order=c(1,0,0))
```

```r
# 상수항
for (t in 2:300){
    y[t] <- -0.8*y[t-1]+e[t]+18  # ø
}

ggtsdisplay(y)
Arima(y, order=c(1,0,0))
```
