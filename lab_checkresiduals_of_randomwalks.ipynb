```r
library(forecast)
library(ggplot2)

set.seed(123)
y <- ts(rnorm(400))
ts_rw <- y

for (t in 2:400){
    y[t] = y[t-1] + ts_rw[t] # random walk process
}

plot(y)
lines(fitted(naive(y)), col="red", lwd=1)
checkresiduals(naive(y))
```

```r
library(forecast)
library(ggplot2)

set.seed(123)
y <- ts(rnorm(400))
ts_rw <- y

for (t in 2:400){
    y[t] = 0.3+y[t-1] + ts_rw[t] # random walk process with drift
}

plot(y)
lines(fitted(naive(y)), col="red", lwd=1)
checkresiduals(naive(y))
```
