# 분해법_추세순환성분 추정을 위한 이동평균방법 적용 실습2

이 실습은 분해법_추세순환성분 추정을 위한 이동평균방법 적용 실습1에 이어서 직접 도식화하여 확인하는 실습입니다.<br/>
M이 3,5,7,9일 때를 각각 확인해보았는데요, M값이 커짐에 따라 가운데로 집약되며, 양끝 데이터가 없어지는 것을 확인할 수 있습니다.<br/>
또한 전반적인 추세만이 남았다는 것도 확인할 수 있습니다.<br/>

분해법 적용을 위한 추세순환성분 추정은 성공적이라는 것을 알 수 있습니다.<br/>

```r
library(fpp2)

# 원본데이터 확인
elecsales
plot(elecsales)
```

```r
# 시계열 분해를 위한 이동평균 적용
for (i in 3:10){
    if (i %% 2){
        elecsales.ma <- ma(elecsales, order=i)
        plot(elecsales, main= paste("M=",i))
        lines(elecsales.ma, col="red")
    }
}
```
