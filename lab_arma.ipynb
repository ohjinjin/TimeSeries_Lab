```r
library(forecast)
library(ggplot2)

set.seed(1234)
arma_1_1 <- arima.sim(model=list(order=c(1,0,1), ar=0.7, ma=-0.2),n=200)
ggtsdisplay(arma_1_1)

arma1=Arima(arma_1_1,order=c(1,0,1))
autoarima1 = auto.arima(arma_1_1)

arma1
autoarima1
```
