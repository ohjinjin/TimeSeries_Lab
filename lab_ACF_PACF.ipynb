```r
library(forecast)
library(fpp2)
library(ggplot2)

# google data의 ACF와 PACF 확인
ggtsdisplay(goog200)
```

```r
# google data의 차분 데이터의 ACF와 PACF를 확인
ggtsdisplay(diff(goog200))
```

```r
# 백색잡음 400개 데이터의 ACF와 PACF를 확인
set.seed(123)
y <- ts(rnorm(400))

ggtsdisplay(y)
```

```r
# 랜덤워크 모형 400개 데이터의 ACF와 PACF를 확인
set.seed(123)
y <- ts(rnorm(400))
ts_rw <- y

for (t in 2:400){
    y[t] = y[t-1] + ts_rw[t] # random walk process
}

ggtsdisplay(y)
```

```r
# 랜덤워크 모형 400개 추세 데이터의 ACF와 PACF를 확인
set.seed(123)
y <- ts(rnorm(400))
ts_rw <- y

for (t in 2:400){
    y[t] = 0.3 + y[t-1] + ts_rw[t] # random walk process with drift
}

ggtsdisplay(y)
```
