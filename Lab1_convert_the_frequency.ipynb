```r
#install.packages("TSA")
#install.packages("forecast")
library(TSA)
library(forecast)

# 월별 자료
data(beersales)
str(beersales)
par(mfrow = c(1,1))
plot(beersales,xlab="Time(Monthly)", ylab="Number of bottles of beer", main="Original beersales")
```

```r
# 월별 자료를 분기별 자료로 변환(합산)
beer_quarterly_sum <- aggregate(beersales, nfrequency= 4, FUN = sum)
par(mfrow = c(1,1))
plot(beer_quarterly_sum, xlab="Time(Quarterly)", ylab="Number of bottles of beer", main="Quarterly sum of aggregation")
```

```r
# 월별 자료를 분기별 자료로 변환(평균)
beer_quarterly_mean <- aggregate(beersales,nfrequency=4, FUN=mean)
par(mfrow = c(1,1))
plot(beer_quarterly_mean, xlab="Time(Quarterly)", ylab="Number of bottles of beer", main="Quarterly mean of aggregation")
```

```r
# 월별 자료를 연도별 자료로 변환(합산)
beer_yearly_sum <- aggregate(beersales, nfrequency=1,FUN = sum)
par(mfrow = c(1,1))
plot(beer_yearly_sum, xlab="Time(Yearly)", ylab="Number of bottles of beer", main="Yearly sum of aggregation")
```

```r
# 월별 자료를 연도별 자료로 변환(평균)
beer_yearly_mean <- aggregate(beersales,nfrequency=1, FUN=mean)
par(mfrow = c(1,1))
plot(beer_yearly_mean, xlab="Time(Yearly)", ylab="Number of bottles of beer", main="Yearly mean of aggregation")
```
