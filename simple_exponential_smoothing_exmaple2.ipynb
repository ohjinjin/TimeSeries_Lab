# 단순지수평활법 실습2_알파값 비교

단순지수 평활법은 시계열 관측값들에 대한 가중평균인 평활값으로 미래의 관측값을 예측합니다.<br/>
단순이동평균법과는 달리 최근 자료에 더 많은 가중값을 부여하는 방식으로 0~1사이의 값을 갖는 알파라는 평활상수에 의해 지수적으로 과거로 갈수록 그 값이 미미해지게됩니다.<br/>

단순지수평활법 실습은 아래와 같은 조건에서 이루어집니다.<br/>
fpp2 패키지의 매뉴얼 링크를 보시려면 [여기](https://cran.r-project.org/web/packages/fpp2/fpp2.pdf)를 눌러주세요.<br/>

* 패키지 fpp2에서 제공하는 oil data 이용
* alpha 값 적용 및 비교 : 0.1, 0.5, 0.9, optimal value 차례로 적용
* ses() 함수이용, h=5까지예측
* 분석은 전체가 아닌 start=1996부터(그때부터의 데이터는 추세나 계절성이 거의 드러나지 않습니다.) : window( 데이터, start=1996)

<br/><br/>


```r
# install.packages("fpp2")

library(fpp2)

# 단순지수평활법
oil
plot(oil)
oildata <- window(oil, start=1996)
plot(oildata) # 원래데이터를 먼저 출력해봅시다.
```

```r
# 0.1, 0.5, 0.9 차례로 알파값을 수정해 결과를 확인해봅니다.
x = c(0.1, 0.5, 0.9)

for (i in x){
    fit_oildata <- ses(oildata, alpha=i,initial="simple", h=5)
    fit_oildata
    plot(fit_oildata)
    lines(fitted(fit_oildata), col="red")
    
    summary(fit_oildata)
    names(fit_oildata)
}
```

```r
# 알파값과 초기값을 최적으로 둡니다.
fit_oildata <- ses(oildata, initial="optimal", h=5) # 생략시 default는 optimal
fit_oildata
plot(fit_oildata)
lines(fitted(fit_oildata), col="red")

summary(fit_oildata)
names(fit_oildata)
```
