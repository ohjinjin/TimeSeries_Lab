```r
library(forecast)
library(fpp2)

austourists_data <- window(austourists, start=2005)
austourists_ets <- ets(austourists_data, model="ZZZ")

summary(austourists_ets)
```

```r
autoplot(austourists_ets)
```

```r
austourists_ets2 <- ets(austourists_data, model="MAM")
summary(austourists_ets2)
```

```r
autoplot(austourists_ets2)
```

ETS(ZZZ)했을 때 ETS(MAM)으로 확인되어 수동으로 MAM 설정 후 적합시켜 확인해보면 모형이 같다는 것을 확인할 수 있습니다.<br/>

무슨 값을 기준으로 특정 메소드를 자동으로 찾아주게 될까요?<br/>
정보손실 기준이라고하는 AIC, AICC, BIC를 기준으로 찾아주게 됩니다.<br/>
이 값들이 작을수록 해당 모형이 선택될 확률이 큽니다.<br/>
이 커리큘럼에서 이내용을 학습하지는 않지만 보통 이 세 값이 작으면 RMSE도 작은 경우도 많습니다<br/>
