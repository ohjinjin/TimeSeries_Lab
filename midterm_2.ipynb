```r
library(forecast)
library(fpp2)

# step1
cangas.ma <- ma(cangas, 12)
plot(cangas)
lines(cangas.ma, col="red", lwd=4)

# step2 ~ step4
# 승법 모형 적용
cangas.mul <- decompose(cangas, type="multiplicative")

# 추세성분 추정
cangas.mul$trend

# 추세조정된 계열: y/T 측정
detrendedseries2 <- cangas.mul$x / cangas.mul$trend
plot(detrendedseries2)
# step2 ~ step4
# 계절성분 추정
cangas.mul$seasonal

# 계절성분 주기별로 다 똑같은 값을 갖으니 그 대표값만 출력
cangas.mul$figure

# 불규칙한 성분 추정
cangas.mul$random

# 동일한 결과
cangas.mul$x / (cangas.mul$trend * cangas.mul$seasonal)

# 차트 그려보기
plot(cangas.mul)

detrendedseries3 <- cangas.mul$x / (cangas.mul$trend * cangas.mul$seasonal)
plot(detrendedseries3)
```
