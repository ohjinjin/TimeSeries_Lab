# 단순지수평활법 최적의 알파값 찾기 실습

forecast 패키지에서는 ses()라는 함수로 단순지수평활을 제공하고 있습니다.
ses()를 호출할 때 alpha값을 비워두면 최적의 alpha를 찾아 적합시키게되는데요, 바로 잔차를 기준으로 최적을 찾게 됩니다.
잔차가 최소가 되게 하는 alpha를 찾아주지요.

이 실습에서는 alpha값을 0.00부터 1.00까지 0.01씩 증가시켜 적합한 후 그 때마다의 RMSE를 측정하고 도식화해봅니다.
그 이후 RMSE가 최소가 되게하는 alpha값을 찾아 다시 적합시켜 최적이 맞는지 비교해봅니다.

```r
library(forecast)
library(fpp2)

# load data
usmelec
autoplot(usmelec)

# apply ses by increaments 0.01
list_rmse <- c()
list_alpha <- c()
tmp_alpha <- 0.00

while (tmp_alpha <= 1.0){
    ses_usmelec <- ses(usmelec, initial="simple", alpha = tmp_alpha)
    list_rmse <- append(list_rmse, sqrt(mean(ses_usmelec$residuals^2)))
    list_alpha <- append(list_alpha, tmp_alpha)
    tmp_alpha <- tmp_alpha + 0.01
    tmp_alpha <- round(tmp_alpha, 2) # to eliminate floating point errors
}

# No3. figure out minimal rmse and optimal alpha
min_rmse_idx <- which.min(list_rmse)
min_rmse <- list_rmse[min_rmse_idx]
opt_alpha <- list_alpha[min_rmse_idx]

min_rmse
opt_alpha

plot(list_alpha, list_rmse, type = "l")

ses_usmelec <- ses(usmelec, initial="simple", alpha = opt_alpha)
ses_optima_usmelec <- ses(usmelec, initial="simple")
summary(ses_usmelec)
summary(ses_optima_usmelec)
```
