```r
library(fpp2)
library(forecast)

oildata <- window(oil, start=1996)
plot(oildata)
```

```r
ses_oildata <- ses(oildata, h = 12)
summary(ses_oildata)
plot(ses_oildata)
```

```r
fit_oildata <- ets(oildata)
summary(fit_oildata)
plot(forecast(fit_oildata, h = 12))
```

```r
plot(oildata, xlim=c(2000,2030), ylim=c(200,800))
lines(simulate(fit_oildata, 12), col="red") #  simulate
```

```r
plot(oildata, xlim=c(2000,2030), ylim=c(200,800))
for (i in 1:5000){ # R 1   .
    lines(simulate(fit_oildata, 12), col="red") # 5000 simulate
}
```
