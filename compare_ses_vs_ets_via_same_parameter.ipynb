```r
library(forecast)
library(fpp2)

# data for ses() vs ets()
oil_data <- window(oil, start=1996)
# check original data
plot(oil_data)

# apply ses
ses_oildata <- ses(oil_data, h = 12, alpha = 0.8339)
summary(ses_oildata)
plot(ses_oildata)
lines(fitted(ses_oildata), type="o", col="red")

# apply ets, ANN
fit_oildata <- ets(oil_data, model= "ANN", alpha = 0.8339, use.initial.values="446.5695")
summary(fit_oildata)
plot(forecast(fit_oildata, h = 12))
lines(fitted(fit_oildata), type="o", col="red")

# apply ets, MNN
fit_oildata <- ets(oil_data, model= "MNN", alpha = 0.8339, use.initial.values="446.5695")
summary(fit_oildata)
plot(forecast(fit_oildata, h = 12))
lines(fitted(fit_oildata), type="o", col="red")
```
