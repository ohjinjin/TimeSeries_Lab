# 시계열분해를 위한 이동평균

이동평균의 이동평균을 이용한 추세성분 추정에 대한 실습 예제입니다.
추세성분 추정 시 m이 짝수로 설정된 경우 중심화 이동평균을 사용하게 됩니다.

```r
library(fpp2)

beer2 <- window(ausbeer, start=1992)
plot(beer2)
ma4 <- ma(beer2, order=4, centre=FALSE)
ma2X4 <- ma(beer2, order=4, centre=TRUE)
plot(beer2)
lines(ma4, col="red")
lines(ma2X4, col="blue")
```
