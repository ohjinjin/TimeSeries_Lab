```r
library(forecast)
library(ggplot2)

x <-rnorm(300)
w <-x

# 0.3 절편, 상승
for(t in 2:300){
    x[t] = 0.3 + x[t-1]+ w[t]
}
x.ts<-ts(x)
ggtsdisplay(x.ts)
ggAcf(x.ts)
ggPacf(x.ts)
```

```r
# -0.4 절편, 하락
for(t in 2:300){
    x[t] = -0.4 + x[t-1]+ w[t]
}
x.ts<-ts(x)    
ggtsdisplay(x.ts)
ggAcf(x.ts)
ggPacf(x.ts)

#a <-ts(c(4, 6, 7, 10))
#diff(a)
#ggtsdisplay(diff(x.ts))
```
