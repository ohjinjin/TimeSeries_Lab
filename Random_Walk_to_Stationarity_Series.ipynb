```r
library(forecast)
library(ggplot2)

# 차분 diff 적용한 예시
a <-ts(c(4, 6, 7, 10))
diff(a)
```

```r
x <-rnorm(300)
w <-x

# 비정상 데이터 확인
for(t in 2:300){
    x[t] = x[t-1]+ w[t]
}
x.ts<-ts(x)
ggtsdisplay(x.ts)
ggAcf(x.ts)
ggPacf(x.ts)
```

```r
# 차분을 적용한 뒤 정상성 확인
ggtsdisplay(diff(x.ts))
ggAcf(diff(x.ts))
ggPacf(diff(x.ts))
```
